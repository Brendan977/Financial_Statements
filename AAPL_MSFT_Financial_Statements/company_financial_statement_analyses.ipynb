# A Tale of Two Tech Giants: 
# Exploring the Financial Evolution of Apple and Microsoft

### **Introduction:**
In the fast-paced world of technology, few companies have had as profound an impact as Apple and Microsoft. Over the past decade, these two tech giants have navigated through changing markets, evolving consumer preferences, and technological innovations. In this analysis, we delve into the financial data of Apple and Microsoft to uncover insights into their growth, profitability, and strategic decisions.


```python
# Dependencies

import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# Create the file path

file_path = Path("Resources/financial_statements.csv")
```

```python
# Read in the financial_statements csv as a dataframe

df = pd.read_csv(file_path)
df
```

### Assess the dataset

```python
df.info()

# NOTE: Notice there is a null value for Market Cap column
```

```python
# Find null row

df.loc[df['Market Cap(in B USD)'].isnull()]

# NOTE: Since this row will not impact the analysis of Apple or Microsoft, we can leave it be.
# Eitherwise it would be best to drop the row using dropna() or fill the cell using fillna() and use the rest of the data for analysis.
```

```python
# Identify all column headers

df.columns

# NOTE: Company column has a whitespace ('Company ')
```

```python
# Create two different dataframes by filtering out all data for apple and all data for microsoft

apple_df = df[df['Company ']=='AAPL']
microsoft_df = df[df['Company ']=='MSFT']
```

```python
# Display apple df

apple_df

# NOTE: NO 2023 data for APPLE
```

```python
# Display microsoft df

microsoft_df
```

###

### The Rise of Apple: From Innovation to Dominance
- **Visual:** Line chart showing Apple's market capitalization over the years.
- **Narrative:** Apple's journey from a niche computer company to a global powerhouse is marked by innovation, iconic product launches, and strategic acquisitions. Despite fluctuations in the market, Apple's market capitalization has steadily risen, reflecting investor confidence in its ability to innovate and dominate multiple sectors, from smartphones to wearables.


```python
# Create a Line chart using matplot for Apple's Market Capitalization through the years 

# identify x and y axes 
year = apple_df['Year']
market_cap = apple_df['Market Cap(in B USD)']

# create plot and formatting
plt.figure(figsize=(10,6))
plt.plot(year, market_cap, marker='o', color='b')
plt.title("Apple's Market Capitalization Over the Years (2009-2022)")
plt.xlabel("Year")
plt.ylabel("Market Cap (Billions)")
plt.grid(True)
plt.xticks(year)
plt.tight_layout()

# save the chart to the images folder
plt.savefig('Images/Apple_Market_Cap.png')

# Display chart
plt.show()
```

###

### Microsoft's Transformation: Reinventing for the Future
- **Visual:** Line chart comparing Microsoft's revenue and net income over the years.
- **Narrative:** Microsoft's transformation under the leadership of Satya Nadella has been remarkable. The company has shifted its focus from traditional software to cloud computing and services, leading to significant revenue growth and improved profitability. Despite facing challenges in the past, Microsoft's strategic pivot has positioned it as a leader in the digital era.


```python
# Create a Line chart using matplot to compare MSFT's revenue and net income over the years

# Plotting subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# identify x and y axes(2) 
year = microsoft_df['Year']
msft_revenue = microsoft_df['Revenue']
msft_net_income = microsoft_df['Net Income']

# Revenue Line Chart
ax1.plot(year, msft_revenue, marker='o', color='green')
ax1.set_title("Microsoft's Revenue Over the Years (2009-2023)")
ax1.set_xlabel('Year')
ax1.set_ylabel('Revenue (Millions)')
ax1.grid(True)

# Net Income Line Chart
ax2.plot(year, msft_net_income, marker='o', color='yellow')
ax2.set_title("Microsoft's Net Income Over the Years (2009-2023)")
ax2.set_xlabel('Year')
ax2.set_ylabel('Net Income (Millions)')
ax2.grid(True)
plt.tight_layout()

# Save line chart to images folder
plt.savefig('Images/MSFT_Revenue_NetIncome.png')

# Display line charts
plt.show()
```
